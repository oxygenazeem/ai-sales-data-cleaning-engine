# AI-driven Sales Data Cleaning & Analytics Engine

## Utilizes Pandas, Numpy, Openpyxl, and G4F

### Phase 1: Environment Setup & Data Generation

```python
import pandas as pd
import numpy as np

data = {
    'Product_ID': [101, 102, np.nan, 104, 105, 106, 107, np.nan],
    'Price': ['120$', '85', 'missing', '45.50', np.nan, '150$', '65', '20'],
    'Quantity': [3, 1, 5, -2, 4, np.nan, 2, 10],
    'Date': ['2026-01-15', '2026/01/16', '01-17-2026', np.nan, '2026-01-19', '2026-01-20', 'unspecified', '2026-01-22']
}

df_messy = pd.DataFrame(data)

df_messy.to_csv("messy_data.csv", index=False)
print("Created messy CSV file successfully!")

df_messy
```

### Phase 2: Data Cleaning & Processing

```python
import pandas as pd
import numpy as np

df = pd.read_csv("messy_data.csv")

# Fixing all prices
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Fix quantity
df.loc[df['Quantity'] < 0, 'Quantity'] = np.nan

df_clean = df.dropna().copy()

df_clean['Product_ID'] = df_clean['Product_ID'].astype(int)

df_clean
```

### Phase 3: Core Statistics Metrics Engine

```python
# Calculate core statistics
total_items_sold = int(df_clean['Quantity'].sum())
average_price = float(df_clean['Price'].mean())
total_revenue = float((df_clean['Price'] * df_clean['Quantity']).sum())

stats_summary = f"""
---DATA ANALYSIS REPORT---
- Total Unique Products Sold: {len(df_clean)}
- Total Units Sold: {total_items_sold}
- Average Product Price: ${average_price:.2f}
- Total Generated Revenue: ${total_revenue:.2f}
"""

print(stats_summary)
```

### Phase 4: G4F AI Layer

```python
import g4f
from g4f.client import Client

#Initialize the AI client
client = Client()

#Crafting a prompt that passes our Python variables straight to the AI model
prompt = f"""
You are an expert financial and business data analyst.
Look at these specific statistics calculate from the sales dataset:

{stats_summary}

Provide a highly concise, professional 3-sentence summary for an executive:
1. One sentence explaining the overall performance (Revenue vs Units sold).
2. One sentence highlighting an observation about the average price.
3. One sentence giving a clear business recommendation based on these numbers.
"""

print("Sending data to AI... (This may take a couple seconds)")

#Request text completion
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": prompt}]
)

#Extract and print the response
ai_analysis = response.choices[0].message.content
print("\nAI Executive Summary")
print(f"\n{ai_analysis}")
```

### Phase 5: Multi-layer Excel Export

```python
#Create an Excel writer object
with pd.ExcelWriter("clean_data_file.xlsx", engine="openpyxl") as writer:
    summary_lines = ai_analysis.split("\n")
    df_summary = pd.DataFrame({"AI Executive Summary": summary_lines})

    df_summary.to_excel(writer, sheet_name="Sales Report", index=False, startrow=0)
    df_clean.to_excel(writer, sheet_name="Sales Report", index=False, startrow=6)

print("Full report saved!")
```

The final result is in the form of an Excel (.xlsx) file, where you can customize the data to your liking!
